### Практическое задание:

###### На основе выбранных продуктов из первого вебинара, выберите 2 A/B-теста и составьте план эксперимента, чтобы избежать проблемы с просмотром (Peeking Problem) 

#### Приоритет гипотезы из второго урока с помощью ICE

### Решение

##### На основе задания из второго урока было принято решение по итогам ICE, предусматривающее шаблоны эксперимента для двух гипотез, набравших большое количество очков:

- Изменить цвет в шапке сайта с чёрного на цветной (синий) для улучшения конверсии - 120 баллов
- Блок «Создать свою сборку» передвинуть левее к блоку «Собрать компьютер» - 100 баллов

### 2.  Изменение цвета шапки сайта.

#### Гипотеза:

Если на странице  https://www.citilink.ru/  изменить цвет в шапке сайта с чёрного на цветной (синий) для улучшения конверсии, то % показателя нахождения клиентов на сайте увеличится на 5%, потому что навигация по сайту будет более прорисованной, что создаёт улучшение по использованию сайта.

##### Что делаем:

**Контрольная версия:** оставляем текущий вид страницы.

**Тестовая версия:**   изменить цвет в шапке сайта с чёрного на цветной (синий)

**На каких пользователях тестируем:** Только на новых пользователях.

###### Метрики:

- Процент конверсии в заказ (%CR) — первичный.
- Активное (вовлечённое) время нахождения на сайте — вторичный.
- Кол-во кликов (переходов по баннерам, просмотров страниц) — вторичный.

###### План действий

- Если наш эксперимент будет применяться и мы зафиксируем преследуемое улучшение в метриках, то масштабируем изменение и «выкатываем» его на всех пользователей.
- Если основные метрики падают или не оправдываются, откатываем эксперимент.
